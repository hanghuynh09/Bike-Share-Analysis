# **Data Importing and Combining**

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import glob
```

```python
def get_csv(file):
    data = pd.read_csv(file)
    return data
```

```python
col=['ride_id','rideable_type','started_at','ended_at','start_station_name','start_station_id','end_station_name'\
                                               ,'end_station_id','start_lat','start_lng','end_lat','end_lng','member_casual']
```

```python
def extract():
    extracted_data = pd.DataFrame(columns=col)
    all_files = glob.glob( "/kaggle/input/bike-share/*.csv")
    for csv in all_files:
        extracted_data = pd.concat([extracted_data, get_csv(csv)], ignore_index=True)
        
        
    return extracted_data
```

```python
df = extract()
```

# **Data Cleaning**

```python
# number of missing values in each column
missing_value = df.isnull()
for column in missing_value.columns.values.tolist():
    print(missing_value[column].value_counts())
    print("")
```

```python
# drop missing values, setting indexes
df.dropna(subset=['start_station_name','start_station_id','end_station_name','end_station_id','end_lat','end_lng'], axis = 0, inplace = True)
df.reset_index(drop=True, inplace=True)
# examine dataframe if there any missing value
missing_value = df.isnull().any()
missing_value
```

```python
# Checking if there are duplicates in dataframe
duplicated_value = df.duplicated().any()
print(duplicated_value)
```

```python
# Checking if there are blank values
blank_value = df.eq('').any()
blank_value
```

```python
# change datatypes of 'started_at' and 'ended_at' 
df['started_at']= pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
df.dtypes
```

```python
# timestamp that little than 1 minute
total_sec = (df.ended_at - df.started_at) / pd.Timedelta(seconds=1)
sec_diff = total_sec.tolist()

# timestamp that more than a day
total_hour = (df.ended_at - df.started_at) / pd.Timedelta(hours=1)
hour_diff = total_hour.tolist()

# add two temporary calculated columns to data
new_data = {"sec_diff":sec_diff, "hour_diff":hour_diff}
df = df.assign(**new_data)

#remove rows with timestamp that don't meet the requirement
df = df.drop(df[(df.sec_diff < 60) | (df.hour_diff > 24)].index)

# remove the temporary columns
df.drop(columns = ['sec_diff','hour_diff'], axis = 1, inplace = True)
df.head()
```

# **Data Transformation**

```python
# add 'ride_length' that calculate bike trip timestamp in minutes
df['ride_length'] = (df['ended_at']-df['started_at'])/pd.Timedelta(minutes=1)
```

```python
# add 'month' columns to indicate month of bike trip
df['month'] = df['started_at'].dt.strftime('%b %y')
df.head()
```

```python
# calculate day of week
df['day_of_week'] = df['started_at'].dt.strftime('%a')
df.head()
```

```python
# round 'ride_length' column to two decimal 
df['ride_length'] = df['ride_length'].round(2)
df.head()
```

```python
# add 'hour_of_day' column
df['hour_of_day'] = df['started_at'].dt.strftime('%I %p')
df.head()
```

# **Data Visualization**

**1. Bike Types Used Per User**

```python
# percentage of cyclist types and rideable type in whole data 
percentage_casual = round(df[df['member_casual'] == 'casual'].shape[0]/df.shape[0],2)
percentage_member = 1- percentage_casual
slice = [percentage_casual,percentage_member]
label = ['casual','member']

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1) 
plt.pie(slice, labels = label, autopct = '%1.0f%%',textprops={'fontsize': 10})
plt.title("Percentage of cyclist type", fontsize = 10, fontweight = 'bold', )

percentage_classic = round(df[df['rideable_type'] == 'classic_bike'].shape[0]/df.shape[0],2)
percentage_electric = round(df[df['rideable_type'] == 'electric_bike'].shape[0]/df.shape[0],2)
percentage_docked = 1 - (percentage_classic + percentage_electric)
slice2 = [percentage_classic,percentage_electric,percentage_docked]
labels2 = ['classic_bike','electric_bike', 'docked_bike']
colors = ['gold', 'lightgreen','lightskyblue']
plt.subplot(1, 2, 2) 
plt.pie(slice2, labels = labels2, autopct = '%1.0f%%', colors=colors, textprops={'fontsize': 10})
plt.title("Percentage of rideable type", fontsize = 10, fontweight = 'bold')

plt.subplots_adjust(wspace=0.5)
plt.show()
```

* **Classic Bikes** : Most popular among all users 
* **Electric Bikes**: Second in popularity
* **Docked Bikes** : Least used, primarily used by casual riders

- Members significantly prefer classic bikes over other types.
- Casual riders use docked bikes exclusively and also have a substantial share of electric bike usage.
- Electric bikes show the most balanced usage between the two rider types.

**2. Rides per Day of Week, Hour of Day and Month**

```python
from matplotlib.ticker import FuncFormatter


def plot_day_hour(df, column_x):
    
     # Define the order for categorical columns
    if column_x == 'day_of_week':
        order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    elif column_x == 'hour_of_day':
        order = ['01 AM', '02 AM', '03 AM', '04 AM', '05 AM', '06 AM', '07 AM', '08 AM', '09 AM', '10 AM', '11 AM', '12 PM',
                 '01 PM', '02 PM', '03 PM', '04 PM', '05 PM', '06 PM', '07 PM', '08 PM', '09 PM', '10 PM', '11 PM', '12 AM']
    elif column_x == 'month':
         order = ['Jul 22', 'Aug 22', 'Sep 22', 'Oct 22', 'Nov 22', 'Dec 22', 'Jan 23', 'Feb 23', 'Mar 23', 'Apr 23', 'May 23', 'Jun 23']
    else:
        print("Invalid column name!")
        return

    df[column_x] = pd.Categorical(df[column_x], categories=order, ordered=True)
    df_grouped = df.groupby([column_x, 'member_casual'], observed=False).size().unstack(fill_value=0).reset_index()
    df_grouped_sorted = df_grouped.sort_values(by=column_x)
    
    plt.figure(figsize=(10,5))
    bar_width = 0.35
    
    r_casual = range(len(df_grouped_sorted))
    r_member = [x + bar_width for x in r_casual]

    # Plotting bars for riders
    bars1 = plt.bar(r_casual, df_grouped_sorted['casual'], width=bar_width, edgecolor='grey', label='Casual')
    bars2 = plt.bar(r_member, df_grouped_sorted['member'], width=bar_width, edgecolor='grey', label='Member')

    plt.xticks([r + bar_width / 2 for r in range(len(df_grouped_sorted))], df_grouped_sorted[column_x], rotation=45, color='gray')
    plt.gca().set_facecolor('white')
   
    # Adjust layout to prevent clipping of labels
    plt.tight_layout()
    
    # Calculate median of ride counts
    median = np.median(df_grouped_sorted[['casual', 'member']].values)
    plt.axhline(median, color='grey', linestyle='--', label='Median')

    # Formatting the y-axis tick labels
    def format_func(value, tick_number):
        if value >= 1000:
            value = int(value) / 1000
            return f'{value:.0f}k'
        else:
            return int(value)
    plt.gca().yaxis.set_major_formatter(FuncFormatter(format_func))
    
    # Function to add labels to bars
    def autolabel(bars):
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width() / 2, height + 1.0, format_func(height, 0), ha='center', va='bottom', fontsize=10, rotation = 90 )
    autolabel(bars1)
    autolabel(bars2)
    
    # Set appropriate title
    if column_x == 'month':
        plt.title('Rides per Month',loc='left', fontweight='bold' , fontsize=12, pad=20)
    elif column_x == 'day_of_week':
        plt.title('Rides per Day of Week',loc='left', fontweight='bold' , fontsize=12, pad=20)
    elif column_x == 'hour_of_day':
        plt.title('Rides per Hour of Day',loc='left', fontweight='bold' , fontsize=12, pad=20)    
    
    # Add vertical dashed lines to separate each month
    for i in range(len(df_grouped_sorted) - 1):
        plt.axvline(x=i + 0.70, color='lightgray')
    # formatting x-axis value style
    if column_x == 'hour_of_day':    
        plt.xticks(color='gray', fontsize = 10, rotation = 90)
    else:
        plt.xticks(color='gray', fontsize = 10, rotation = 45)
    
    plt.legend(loc="upper right",  bbox_to_anchor=(1,1), fontsize=8 )
    plt.show()

plot_day_hour(df, 'day_of_week')  
```

- Weekdays:

Members consistently ride more than casual riders during the weekdays. This could be indicative of regular commuting patterns.
The lowest member participation is on Monday, with 366,000 rides, and it steadily increases towards Thursday and Friday.

- Weekends:

There is a notable surge in casual rider participation on weekends, especially on Saturday, where the rides by casual riders almost match those of members.
Sunday's rides are also higher for casual riders compared to most weekdays, suggesting leisure or recreational usage over the weekend.

```python
def line_plot(df,column_z):

    df_gr = df.groupby([column_z,'member_casual'], observed=False )['ride_length'].mean().unstack(fill_value=0).reset_index()
    df_gr_sort = df_gr.sort_values(by=column_z) 

    x1 = df_gr_sort['casual']
    x2 = df_gr_sort['member']
    y = df_gr_sort[column_z]

    # Find the indices of the min and max values
    marker1 = [x1.idxmin(), x1.idxmax()]
    marker2 = [x2.idxmin(), x2.idxmax()]


    plt.figure(figsize=(10,1.5))
    plt.gca().set_facecolor('white')
    plt.plot(y,x1,marker='o',markevery=marker1, label='casual')
    plt.plot(y,x2,marker='o',markevery=marker2 ,label = 'member')

    # Set appropriate title
    if column_z == 'month':
        plt.title('Average Ride Duration per Month', fontweight='bold' , fontsize=10, pad=20)
    elif column_z == 'day_of_week':
        plt.title('Average Ride Duration per Day of Week', fontweight='bold' , fontsize=10, pad=20)
    elif column_z == 'hour_of_day':
        plt.title('Average Ride Duration per Hour of Day', fontweight='bold' , fontsize=10, pad=20)    
    
    
    # Add labels for each marker
    for i in marker1:
        plt.text(y[i], x1[i] + 0.5, f'{x1[i]:.1f} Min', ha='center', va='bottom', fontsize=10)

    for i in marker2:
        plt.text(y[i], x2[i] + 0.5, f'{x2[i]:.1f} Min', ha='center', va='bottom', fontsize=10)

    if column_z == 'hour_of_day':    
        plt.xticks(color='gray', fontsize = 10, rotation = 90)
    else:
        plt.xticks(color='gray', fontsize = 10)
    plt.gca().get_yaxis().set_visible(False)    
    plt.legend(['casual', 'member'], fontsize=8, loc="upper right",  bbox_to_anchor=(1.1,1))

line_plot(df, 'day_of_week')
line_plot(df, 'hour_of_day')    
line_plot(df, 'month')
```


Findings From Average Ride

- Duration:

    Casual riders cycle longer on average.
    Member journey length remains consistent.
    
- Variations in casual rider journey length:
    Spring and Summer, Weekends.
    10 am to 3 pm during the day & Short trips between 5 and 8 am.

Key Takeaway:
Casual commuters cover approximately twice the distance but ride less frequently compared to members.
Longer rides are prevalent during weekends and daytime hours (outside of typical commuting times)

```python
from matplotlib.ticker import FuncFormatter
def hor_bar(df, column_f):
    df_g = df.groupby([column_f,'member_casual']).size().unstack(fill_value=0).reset_index()
    df_g['total'] = df_g['casual'] + df_g['member']
    df_g_s = df_g.sort_values(by='total', ascending = True).tail(5)
    x1 = df_g_s['casual']
    x2 = df_g_s['member']
    y = df_g_s[column_f]
    plt.figure(figsize=(10,2))
    bar1 = plt.barh(y, x1, label='casual')
    bar2 = plt.barh(y, x2,left=x1, label = 'member')
    
    # Set title to graphs
    if column_f == 'start_station_name':
        plt.title('Top 5 Start Station Names',fontweight='bold', fontsize = 15, loc='left')
    elif column_f == 'end_station_name':
        plt.title('Top 5 End Station Names',fontweight='bold', fontsize = 15, loc='left')
    else:
        print("Invalid column name!")
        return
    

    # Formatting xticks label to 00k
    def format_func(value, tick_number):
        if value >= 1000:
            value = int(value) / 1000
            return f'{value:.0f}k'
        else:
            return int(value)

    def autolabels(bars, offset):
        for bar in bars:
            width = bar.get_width()
            plt.text(bar.get_x() + width / 2, bar.get_y() + bar.get_height() / 2,
                     format_func(width, 0),
                     ha='center', va='center', color='white', fontsize=10)

    autolabels(bar1, 1500)
    autolabels(bar2, 1500)
    
    plt.legend(loc='lower right')
    plt.gca().set_facecolor('white')
    plt.gca().get_xaxis().set_visible(False)
    plt.gca().xaxis.set_major_formatter(FuncFormatter(format_func))

hor_bar(df,'start_station_name')
hor_bar(df,'end_station_name')
```

        Commonly start trips near Museums, Parks, Beaches.

        Frequently end trips near recreational sites.

    Members:

        Often begin trips near Universities, Residential areas, Restaurants, Grocery stores, Schools, Banks, Train stations.
        
        Tend to end trips near universities, residential, and commercial areas.
